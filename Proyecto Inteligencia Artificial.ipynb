🛑 **CARACTERÍSTICAS DEL ACCIDENTE:**
* Accident_Index: Identificador único del accidente, usado como clave principal.

* Location_Easting_OSGR / Location_Northing_OSGR: Coordenadas en el sistema de referencia británico (Ordnance Survey Grid Reference).

* Longitude / Latitude: Coordenadas geográficas del accidente en formato decimal (WGS84).

* LSOA_of_Accident_Location: Código del área estadística local (LSOA) donde ocurrió el accidente.

* Accident_Severity: Gravedad del accidente (1 = Fatal, 2 = Grave, 3 = Leve).

* Number_of_Vehicles: Número de vehículos implicados en el accidente.

* Number_of_Casualties: Número total de víctimas (heridos y/o fallecidos).

* Date: Fecha del accidente (formato día/mes/año).

* Day_of_Week: Día de la semana (1 = Domingo, 2 = Lunes, ..., 7 = Sábado).

* Time: Hora del accidente.

* Year: Año en que ocurrió el accidente.

🏛 **AUTORIDADES:**

* Local_Authority_(District): Autoridad local del distrito donde ocurrió el accidente.

* Police_Force: Código numérico de la fuerza policial que gestionó el accidente.

* Local_Authority_(Highway): Autoridad responsable de la vía donde ocurrió el accidente.

* Did_Police_Officer_Attend_Scene_of_Accident: Indica si un oficial de policía acudió a la escena del accidente ("Yes", "No", "Unknown").

🚧 **CARACTERÍSTICAS VIALES:**
* Urban_or_Rural_Area: Área donde ocurrió el accidente (1 = Urbana, 2 = Rural).

* Carriageway_Hazards: Peligros presentes en la vía (como vehículos descompuestos, objetos, animales, etc.).

* Special_Conditions_at_Site: Condiciones especiales en el lugar del accidente (desvíos, obras, etc.).

* Road_Surface_Conditions: Estado de la superficie de la carretera (seca, mojada, con hielo, etc.).

* 1st_Road_Class / 1st_Road_Number: Clase y número de la vía principal (A, B, C, etc.).

* 2nd_Road_Class / 2nd_Road_Number: Clase y número de la vía secundaria en intersecciones.

* Road_Type: Tipo de vía (recta, intersección, rotonda, etc.).

* Speed_limit: Límite de velocidad en el lugar del accidente (en mph).

* Junction_Detail: Detalles del cruce (sin cruce, T, Y, rotonda, etc.).

* Junction_Control: Control en la intersección (semaforizado, sin control, etc.).

* Pedestrian_Crossing-Human_Control: Control humano en cruces peatonales (guardias escolares, semáforos operados manualmente, etc.).

* Pedestrian_Crossing-Physical_Facilities: Infraestructura física de cruces peatonales (paso de cebra, semáforo, puente, etc.).

☁️ **CONDICIONES AMBIENTALES:**
* Light_Conditions: Condiciones de luz al momento del accidente (día, noche con o sin luz artificial, etc.).

* Weather_Conditions: Condiciones meteorológicas (lluvia, nieve, niebla, viento fuerte, etc.).


```python
#@title Visualizacion de datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 10)


fig, axs = plt.subplots(2, 2)


sns.countplot(data=df, x='Accident_Severity', palette='viridis', ax=axs[0, 0])
axs[0, 0].set_title("Distribución de la Gravedad del Accidente")
axs[0, 0].set_xlabel("1=Fatal, 2=Grave, 3=Leve")
axs[0, 0].set_ylabel("Número de Accidentes")

sns.scatterplot(data=df, x='Number_of_Vehicles', y='Number_of_Casualties', hue='Accident_Severity', palette='deep', ax=axs[0, 1], alpha=0.5)
axs[0, 1].set_title("Relación Vehículos-Víctimas")
axs[0, 1].set_xlabel("Número de Vehículos")
axs[0, 1].set_ylabel("Número de Víctimas")


top_weather = df['Weather_Conditions'].value_counts().nlargest(10).index  # Top 10
sns.countplot(data=df[df['Weather_Conditions'].isin(top_weather)], y='Weather_Conditions', palette='magma', ax=axs[1, 0])
axs[1, 0].set_title("Accidentes según Condiciones Meteorológicas (Top 10)")
axs[1, 0].set_xlabel("Número de Accidentes")
axs[1, 0].set_ylabel("Condiciones Meteorológicas")


corr_matrix = df[['Accident_Severity', 'Number_of_Vehicles', 'Number_of_Casualties', 'Speed_limit']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title("Mapa de Correlación")


plt.tight_layout()
plt.show()
```

```python
#@title porcentaje de desbalanceo para cada columna
desbalanceo = {}
for col in df.columns:
    # Contar la frecuencia de cada valor en la columna, excluyendo NaN
    value_counts = df[col].value_counts(normalize=True, dropna=True)
    if not value_counts.empty:
        # El desbalanceo se puede considerar como 1 menos la frecuencia del valor más común
        desbalanceo[col] = (1 - value_counts.max()) * 100
    else:
        # Si la columna no tiene valores no-NaN, el desbalanceo es 0
        desbalanceo[col] = 0.0

# Convertir el diccionario de desbalanceo a DataFrame
df_desbalanceo = pd.DataFrame(list(desbalanceo.items()), columns=['Nombre Columna', 'Porcentaje Desbalanceo'])

# Ordenar el DataFrame por porcentaje de desbalanceo de forma descendente
df_desbalanceo_ordenado = df_desbalanceo.sort_values(by='Porcentaje Desbalanceo', ascending=False)

print("\nPorcentaje de desbalanceo de cada columna (ordenado por desbalanceo):")
display(df_desbalanceo_ordenado)
```

# Segunda entrega - Modelos Supervisados



```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import pandas as pd
from imblearn.over_sampling import SMOTE

df = pd.read_csv("accidents_2012_to_2014.csv", low_memory=False)

cols_to_drop = [
    'Accident_Index', 'Date', 'Time', 'LSOA_of_Accident_Location',
    'Local_Authority_(District)', 'Local_Authority_(Highway)',
    'Location_Easting_OSGR', 'Location_Northing_OSGR',
    'Longitude', 'Latitude', 'Junction_Detail'
]
df_clean = df.drop(columns=cols_to_drop)
df_clean = df_clean.fillna(df_clean.mode().iloc[0])


# Codificación one-hot para las variables categóricas
df_encoded = pd.get_dummies(df_clean, drop_first=True)

X = df_encoded.drop('Accident_Severity', axis=1)
y = df_encoded['Accident_Severity']

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)


rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("\n Random Forest:")
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))


lr = LogisticRegression(max_iter=500, random_state=42)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("\n Logistic Regression:")
print(classification_report(y_test, y_pred_lr))
print("Accuracy:", accuracy_score(y_test, y_pred_lr))


#svm = SVC(random_state=42)
#svm.fit(X_train, y_train)
#y_pred_svm = svm.predict(X_test)
#print("\n Support Vector Machine:")
#print(classification_report(y_test, y_pred_svm))
#print("Accuracy:", accuracy_score(y_test, y_pred_svm))

```

```python
#@title Redes Neuronales

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Escalar los datos para mejorar el rendimiento de la red
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_res)


y_cat = to_categorical(y_res - 1)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_cat, test_size=0.2, random_state=42, stratify=y_cat)

# Definir el modelo
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_data=(X_test, y_test), verbose=1)


loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f" Precisión final de la red neuronal: {accuracy * 100:.2f}%")
```
